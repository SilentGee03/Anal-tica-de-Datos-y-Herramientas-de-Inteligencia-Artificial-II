```python
# Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Datos de Facturación.xlsx')
# df
```

```python
# Información del DataFrame
df.info()
```

```python
# Imprimo los primeros 5 registro del dataframe
df.head()
```

```python
# Convertir la columna CVE_CLPV a numérico, forzando errores a NaN si es necesario
df['CVE_CLPV'] = pd.to_numeric(df['CVE_CLPV'], errors='coerce')
# Lo intenté con el astype(int) pero me marcaba error
```

```python
# Filtro 1
filtro1=df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
filtro1
```

```python
# Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro1.csv")
```

```python
# Filtro 2
filtro2 = df[~df['CVE_VEND'].isin([5, 4])]
filtro2
```

```python
# Convertir archivo filtrado a CSV
filtro2.to_csv("Filtro2.csv")
```

```python
# Filtro 3
filtro3 = df[df['FECHA_ENT'] == '28/02/2022  12:00:00 a. m.']
filtro3
```

```python
# Convertir archivo filtrado a CSV
filtro3.to_csv("Filtro3.csv")
```

```python
# Filtro 4
filtro4 = df[(df['CAN_TOT'] < 5951.7) | (df['STATUS'] == 'E')]
filtro4
```

```python
# Convertir archivo filtrado a CSV
filtro4.to_csv("Filtro4.csv")
```

```python
# Filtro 5
filtro5 = df[['CVE_DOC', 'FECHA_ENT', 'FECHA_VEN', 'CAN_TOT']]
filtro5
```

```python
# Convertir archivo filtrado a CSV
filtro5.to_csv("Filtro5.csv")
```

```python
# Filtro 6
filtro6 = df.iloc[7000:7099]  # Filas 7001 a 7099
filtro6
```

```python
# Convertir archivo filtrado a CSV
filtro6.to_csv("Filtro6.csv")
```

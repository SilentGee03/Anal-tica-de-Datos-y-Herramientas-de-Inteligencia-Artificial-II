```python
# Importar las librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
import ast
```

## Etapa 1

```python
#Realizamos una copia del dataframe
data2=dfSeattle.copy()
```

```python
# Eliminamos los símbolos que hacen que la columna de precio sea de tipo objeto
dfSeattle['host_response_rate'] = dfSeattle['host_response_rate'].str.replace('%', '')
dfSeattle['host_acceptance_rate'] = dfSeattle['host_acceptance_rate'].str.replace('%', '')
dfSeattle['host_is_superhost'] = dfSeattle['host_is_superhost'].str.replace('f', '0')
dfSeattle['host_is_superhost'] = dfSeattle['host_is_superhost'].str.replace('t', '1')
dfSeattle['host_has_profile_pic'] = dfSeattle['host_has_profile_pic'].str.replace('f', '0')
dfSeattle['host_has_profile_pic'] = dfSeattle['host_has_profile_pic'].str.replace('t', '1')
dfSeattle['host_identity_verified'] = dfSeattle['host_identity_verified'].str.replace('f', '0')
dfSeattle['host_identity_verified'] = dfSeattle['host_identity_verified'].str.replace('t', '1')
dfSeattle['instant_bookable'] = dfSeattle['instant_bookable'].str.replace('f', '0')
dfSeattle['instant_bookable'] = dfSeattle['instant_bookable'].str.replace('t', '1')
# Convertir a numérico
dfSeattle['host_response_rate'] = pd.to_numeric(dfSeattle['host_response_rate'], errors='coerce')
dfSeattle['host_acceptance_rate'] = pd.to_numeric(dfSeattle['host_acceptance_rate'], errors='coerce')
dfSeattle['host_is_superhost'] = pd.to_numeric(dfSeattle['host_is_superhost'], errors='coerce')
dfSeattle['host_has_profile_pic'] = pd.to_numeric(dfSeattle['host_has_profile_pic'], errors='coerce')
dfSeattle['host_identity_verified'] = pd.to_numeric(dfSeattle['host_identity_verified'], errors='coerce')
dfSeattle['instant_bookable'] = pd.to_numeric(dfSeattle['instant_bookable'], errors='coerce')
dfSeattle['price'] = dfSeattle['price'].str.replace(r'[\$,]', '', regex=True).astype(float)
```

```python
# Extraemos el año de la última review y la del año donde inició el host de cada lugar
dfSeattle["last_review"]
dfSeattle["host_since"]
# Usar una expresión regular para extraer el año
dfSeattle['last_review'] = dfSeattle['last_review'].str.extract(r'(\d{4})')
dfSeattle['last_review'] = pd.to_numeric(dfSeattle['last_review'], errors='coerce')
dfSeattle['host_since'] = dfSeattle['host_since'].str.extract(r'(\d{4})')
dfSeattle['host_since'] = pd.to_numeric(dfSeattle['host_since'], errors='coerce')
# Agregamos una nueva columna
dfSeattle['host_years'] = dfSeattle["last_review"].max() - dfSeattle['host_since']
```

```python
# Borramos las cualitativas
dfSeattle.drop(["description","neighborhood_overview","host_about","host_neighbourhood","name","neighbourhood","host_location","first_review","has_availability"], axis = 1, inplace = True)

# Borramos las cuantitativas
dfSeattle.drop(["scrape_id","minimum_nights","maximum_nights","minimum_minimum_nights","minimum_maximum_nights", "maximum_minimum_nights", "maximum_maximum_nights","availability_60", "availability_90", "calculated_host_listings_count_entire_homes", "calculated_host_listings_count_private_rooms", "calculated_host_listings_count_shared_rooms","last_review"], axis = 1, inplace = True)
```

```python
# Dividimos por valores cuantitativos y cualitativos
cuantitativas = dfSeattle.select_dtypes(np.number)
cualitativas = dfSeattle.select_dtypes(include=['object'])
```

## Limpieza con los valores cualitativos

## Limpieza con los valores cuantitativos

```python
# Copia
dfLimpiezaCuanti = cuantitativas.copy()
```

```python
# Identificar valores cualitativos nulos
dfLimpiezaCuanti.isnull().sum()
```

```python
cuantitativas[["host_response_rate","host_acceptance_rate"]].head(5)
# Dado que está en porcentaje, se sustituye con media
dfLimpiezaCuanti["host_response_rate"] = dfLimpiezaCuanti["host_response_rate"].fillna(round(cuantitativas["host_response_rate"].mean(),1))
dfLimpiezaCuanti["host_acceptance_rate"] = dfLimpiezaCuanti["host_acceptance_rate"].fillna(round(cuantitativas["host_acceptance_rate"].mean(),1))
```

```python
cuantitativas["host_is_superhost"].value_counts()
# Reemplazaremos por un valor nuevo
dfLimpiezaCuanti["host_is_superhost"] = dfLimpiezaCuanti["host_is_superhost"].fillna(2)
```

```python
#Para valores como baños, habitaciones y camas, estas serán reemplazados por la mediana para no alterar la distribución de los datos
dfLimpiezaCuanti["bathrooms"] = dfLimpiezaCuanti["bathrooms"].fillna(int(round(cuantitativas["bathrooms"].median(), 0)))
dfLimpiezaCuanti["bedrooms"] = dfLimpiezaCuanti["bedrooms"].fillna(int(round(cuantitativas["bedrooms"].median(), 0)))
dfLimpiezaCuanti["beds"] = dfLimpiezaCuanti["beds"].fillna(int(round(cuantitativas["beds"].median(), 0)))
```

```python
"""Ahora vemos la distribución del dataframe con aquellas columnas iniciando con 'review_scores', los valores son escala Likert"""

dfReviewScores = dfLimpiezaCuanti.loc[:, dfLimpiezaCuanti.columns.str.contains("review_scores")]
[dfReviewScores.min(), dfReviewScores.max()]
```

```python
# Sustituidos por la media

dfLimpiezaCuanti["review_scores_rating"] = dfLimpiezaCuanti["review_scores_rating"].fillna(round(cuantitativas["review_scores_rating"].mean(), 2))
dfLimpiezaCuanti["review_scores_accuracy"] = dfLimpiezaCuanti["review_scores_accuracy"].fillna(round(cuantitativas["review_scores_accuracy"].mean(), 2))
dfLimpiezaCuanti["review_scores_cleanliness"] = dfLimpiezaCuanti["review_scores_cleanliness"].fillna(round(cuantitativas["review_scores_cleanliness"].mean(), 2))
dfLimpiezaCuanti["review_scores_checkin"] = dfLimpiezaCuanti["review_scores_checkin"].fillna(round(cuantitativas["review_scores_checkin"].mean(), 2))
dfLimpiezaCuanti["review_scores_communication"] = dfLimpiezaCuanti["review_scores_communication"].fillna(round(cuantitativas["review_scores_communication"].mean(), 2))
dfLimpiezaCuanti["review_scores_location"] = dfLimpiezaCuanti["review_scores_location"].fillna(round(cuantitativas["review_scores_location"].mean(), 2))
dfLimpiezaCuanti["review_scores_value"] = dfLimpiezaCuanti["review_scores_value"].fillna(round(cuantitativas["review_scores_value"].mean(), 2))
```

```python
# Redondeados a valor entero

dfLimpiezaCuanti["review_scores_rating"] = dfLimpiezaCuanti["review_scores_rating"].round()
dfLimpiezaCuanti["review_scores_accuracy"] = dfLimpiezaCuanti["review_scores_accuracy"].round()
dfLimpiezaCuanti["review_scores_cleanliness"] = dfLimpiezaCuanti["review_scores_cleanliness"].round()
dfLimpiezaCuanti["review_scores_checkin"] = dfLimpiezaCuanti["review_scores_checkin"].round()
dfLimpiezaCuanti["review_scores_communication"] = dfLimpiezaCuanti["review_scores_communication"].round()
dfLimpiezaCuanti["review_scores_location"] = dfLimpiezaCuanti["review_scores_location"].round()
dfLimpiezaCuanti["review_scores_value"] = dfLimpiezaCuanti["review_scores_value"].round()
```

```python
# Comprobación del cambio

dfReviewScores = dfLimpiezaCuanti.loc[:, dfLimpiezaCuanti.columns.str.contains("review_scores")]
dfReviewScores.value_counts()
```

```python
# Reviews per month sustituido con la mediana
dfLimpiezaCuanti["reviews_per_month"] = dfLimpiezaCuanti["reviews_per_month"].fillna(round(cuantitativas["reviews_per_month"].median(), 2))
```

```python
# Utilizaremos la mediana para no afectar la distribución de los datos
dfLimpiezaCuanti["price"] = dfLimpiezaCuanti["price"].fillna(round(cuantitativas["price"].median(), 0))
```

```python
# Corroboramos que no tengamos valores nulos
dfLimpiezaCuanti.isnull().sum().sum()
```

## Categorización de valores

```python
# Unimos ambas limpiezas (Valores cuantitativos y cualitativos limpios)
dfSeattleLimpio = pd.concat([cualiLimpieza, dfLimpiezaCuanti], axis=1)
```

```python
# Comprobamos que sí estén todos los datos en ambas bases, tanto la inicial como la que tuvo un proceso de limpieza
dfSeattle.info()
dfSeattleLimpio.info()
```

```python
# Asignamos categorías a las columnas de tipo texto con valores numéricos usando la libreria de sklearn
labels = LabelEncoder()
# Convertir la columna de texto a valores numéricos (Categorizar con números)
dfSeattleLimpio['room_type_num'] = labels.fit_transform(dfSeattleLimpio['room_type'])
dfSeattleLimpio['neighbourhood_cleansed_num'] = labels.fit_transform(dfSeattleLimpio['neighbourhood_cleansed'])
dfSeattleLimpio['host_response_time_num'] = labels.fit_transform(dfSeattleLimpio['host_response_time'])
```

```python
# Obtenemos categorías para bathrooms_text:
dfSeattleLimpio['bathroom_text_clean'] = dfSeattleLimpio['bathrooms_text'].str.replace(r'^\d*\.?\d*\s?', '', regex=True).str.strip()
dfSeattleLimpio['bathroom_text_clean'] = dfSeattleLimpio['bathroom_text_clean'].str.replace("baths","bath")
dfSeattleLimpio['bathroom_text_clean'] = dfSeattleLimpio['bathroom_text_clean'].str.replace("shared baths","shared bath")
dfSeattleLimpio['bathroom_text_clean'].value_counts()
# Las volvemos numéricas:
dfSeattleLimpio['bathroom_text_num'] = labels.fit_transform(dfSeattleLimpio['bathroom_text_clean'])
dfSeattleLimpio['bathroom_text_num'].value_counts()
```

```python
# Conteo para la variable "amenities" y "host_verifications"
dfSeattleLimpio['amenities_num'] = dfSeattleLimpio['amenities'].apply(ast.literal_eval)
dfSeattleLimpio["host_verifications_num"] = dfSeattleLimpio['host_verifications'].apply(ast.literal_eval)
# Crear una nueva columna que cuenta los elementos de la lista
dfSeattleLimpio['amenities_num'] = dfSeattleLimpio['amenities_num'].apply(len)
dfSeattleLimpio['host_verifications_num'] = dfSeattleLimpio['host_verifications_num'].apply(len)
```

```python
# Estandarización a precio mexicano
pesoMex = 19.61
dfSeattleLimpio["priceMex"] = dfSeattleLimpio["price"] * pesoMex
```

```python
# Redondeamos a dos decimales
dfSeattleLimpio["priceMex"] = np.round(dfSeattleLimpio["priceMex"],2)
```

```python
# Corroboramos la conversión

dfSeattleLimpio[["priceMex","price"]].min()
```

## Categorización de la columna precio (Estandarizada a Mx)

```python
#Calculamos el número total de la población "n"
n = len(dfSeattleLimpio['priceMex'])
n
# Límites
Max = dfSeattleLimpio["priceMex"].max()
Min = dfSeattleLimpio["priceMex"].min()
Limites = [Min, Max]
Limites
```

```python
# Calculamos el rango R
R = Max - Min
R
```

```python
# Regla de Sturges: Calcular el número de Intervalos de Clase "ni"
ni = 1 + 3.32*np.log10(n)
ni = int(ni)
ni
```

```python
# Calculamos el ancho del intervalo "i"
i = R/ni
```

```python
# Crear 13 bordes (12 intervalos)
intervalos = np.linspace(Min, Max, 12)
```

```python
# Definir las categorías
categorias = ['Muy Económico', 'Básico', 'Económico', 'Estándar', 'Costo Moderado',
              'Costo Elevado', 'Premium', 'De Lujo', 'Elite', 'Exclusivo', 'Top de Mercado']

# Se asignan en una nueva columna
dfSeattleLimpio['priceMexCategory'] = pd.cut(x = dfSeattleLimpio['priceMex'], bins = intervalos, labels = categorias, include_lowest=True)
```

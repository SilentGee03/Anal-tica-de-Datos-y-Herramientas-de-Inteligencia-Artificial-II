```python
# Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Datos de Facturación.xlsx')
# df
```

```python
# Información del DataFrame
df.info()
```

```python
# Imprimo los primeros 5 registro del dataframe
df.head()
```

```python
# Conversión de columna CVE_CLPV
df['CVE_CLPV'] = df['CVE_CLPV'].astype(int)
```

```python
# Reemplazo de datos tipo string a datos tipo numérico
df.CVE_CLPV = df.CVE_CLPV.replace({'8440Q':'8440'}, regex = False)
```

```python
# Filtro 1
filtro1=df[(df['CVE_CLPV'] >= 1000) & (df['CVE_CLPV'] <= 2000)]
filtro1
```

```python
# Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro1.csv")
```
